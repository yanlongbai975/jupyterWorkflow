Create a new column based on some existing columns

```python
df = pd.DataFrame(
    [
        (1, 'Hello', 158, True, 12.8),
        (2, 'Hey', 567, False, 74.2),
        (3, 'Hi', 123, False, 1.1),
        (4, 'Howdy', 578, True, 45.8),
        (5, 'Hello', 418, True, 21.1),
        (6, 'Hi', 98, False, 98.1),
    ],
    columns=['colA', 'colB', 'colC', 'colD', 'colE']
)
df
```

Method 1: Using apply() method

```python
#Define my own methods
def categorise(row):  
    if row['colC'] > 0 and row['colC'] <= 99:
        return 'A'
    elif row['colC'] > 100 and row['colC'] <= 199:
        return 'B'
    elif row['colC'] > 200  and row['colC'] <= 299:
        return 'C'
    return 'D'
```

```python
df['colF'] = df.apply(lambda row: categorise(row), axis=1)
df
```

```python
#Or just using lambda expression
df['colG'] = df.apply(lambda row: row.colC + row.colE, axis=1)
df
```

Method 2: Using select() method

```python
conditions = [
  np.logical_and(df['colC'].gt(0), np.less_equal(df['colC'], 99)),
  np.logical_and(df['colC'].gt(100), np.less_equal(df['colC'],199)),
  np.logical_and(df['colC'].gt(200), np.less_equal(df['colC'],299)),
]
outputs = ['A', 'B', 'C']
```

```python
df['colH'] = pd.Series(np.select(conditions, outputs, 'D'))
df
```

Method 3: Using loc property

```python
df.loc[
 np.logical_and(df['colC'].gt(0), np.less_equal(df['colC'], 99)), 
 'colI'
] = 'A'

df.loc[
 np.logical_and(df['colC'].gt(100), np.less_equal(df['colC'], 199)),'colI'
] = 'B'

df.loc[
 np.logical_and(df['colC'].gt(200), np.less_equal(df['colC'], 299)),'colI'
] = 'C'

df['colI'].fillna('D', inplace=True)

df
```
