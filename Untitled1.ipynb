```python
def stack(deque, ord, ls):
    if(ord == 0 or ord == 1): 
        if(len(deque)==0):
            print(ls)
            return 
        elif(len(ls) == 0 or deque[0] <= ls[-1]):
            ls.append(deque.popleft())
            ord = 0
            stack(deque, ord, ls)
        else:
            ord += 1
            deque.reverse()
            stack(deque, ord, ls)
    else:
        print(ls)
        ls.clear()
        return 
                        
               
    

import collections
T = int(input())
for i in range(T):
    n = int(input())
    d = collections.deque(list(map(int, input().split(" "))))
    ls = []
    stack(d, 0, ls)
    if(len(ls) == n):
        print("Yes")
    else:
        print("No")
```
